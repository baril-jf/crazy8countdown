```python
#!/usr/bin/env python
# coding: utf-8

# In[85]:


# Prénom, Nom, Matricule
# Prénom, Nom, Matricule
import math
import random
import copy

class LinkedList:
    class _Node:
        def __init__(self, v, n):
            self.value = v
            self.next = n

    def __init__(self):
        self._head = None
        self._size = 0

    def __str__(self):
        if self.isEmpty():
            return "[]"
        else:
            #sinon on passe a travers tous les elements
            liste = "["
            current = self._head
            while current.next != None:
                liste += str(current.value) + ", "
                current = current.next
            liste += str(current.value) + "]" 
            
        return liste
                
    def __len__(self):
        return self._size

    def isEmpty(self):
        return self._size == 0

    # Adds a node of value v to the beginning of the list
    def add(self, v):
        if self.isEmpty():
            self._head = self._Node(v, None)
        else :
            self._head = self._Node(v, self._head)
        
        self._size += 1
        
    # Adds a node of value v to the end of the list
    def append(self,v):
        newNode = self._Node(v, None)
        if self.isEmpty():
            self._head = newNode
        else: 
            current = self._head
            while current.next != None :
                current = current.next
            current.next = newNode    
        self._size += 1

    # Removes and returns the first node of the list
    def pop(self):
        if not self.isEmpty():
            temp = self._head
            self._head = temp.next
            self._size -= 1
            return temp.value
        else: return None

    # Returns the value of the first node of the list
    def peek(self):
        if not self.isEmpty():
            return self._head.value
        else: return None
    # Removes the node of the list with value v and return v
    def remove(self,v):
        prev = None
        current = self._head
        while current:
            if current.value == v:
                if prev:
                    prev.next = current.next
                else:
                    self._head = current.next
                self._size -= 1
                return v
            
            prev= current
            current = current.next
        return None       
        
        
class CircularLinkedList(LinkedList):
    def __init__(self):
        super().__init__()
         
    def __str__(self):
        if self.isEmpty():
            return '[]'
        else:
            #sinon on passe a travers tous les elements
            liste = "["
            current =self._head
            while current.next != self._head:
                liste += str(current.value) + ", "
                current = current.next
            liste += str(current.value) + "]"
        return liste

    def __iter__(self):
        # Iterate the list
        if self.isEmpty():
            yield
        else:
            current = self._head
            while current.next != self._head :
                yield current.value
                current = current.next
            yield current.value

    # Moves head pointer to next node in list
    def next(self):
        if self.isEmpty() :
            return None
        else :
            self._head = self._head.next
        

    # Adds a node of value v to the end of the list
    def append(self, v):
        newNode = self._Node(v, None)
        
        if self.isEmpty():
            self._head = newNode
            self._head.next = self._head
        else:
            current = self._head
            while current.next != self._head:
                current = current.next
            current.next = newNode
            current.next.next = self._head
        self._size += 1

    # Reverses the next pointers of all nodes to previous node
    def reverse(self):
        if self.isEmpty():
            return None
                 
        current = self._head
        prev = None
        suiv = current.next
            
        while suiv != self._head:
            prev = current
            current = suiv
            suiv = suiv.next
            current.next = prev
        suiv.next = current
        self._head = suiv
        
    # Removes head node and returns its value
    def pop(self):
        if self.isEmpty():
            return None
        elif self._size == 1:
            self._size -= 1
            temp = self._head.value
            self._head = None
            return temp
        else :
            temp = self._head
            while temp.next != self._head:
                temp = temp.next
            v = self._head.value
            temp.next = self._head.next
            self._head = temp.next
            self._size -= 1 
            return v
class Card:
    def __init__(self, r, s):
        self._rank = r
        self._suit = s

    suits = {'s': '\U00002660', 'h': '\U00002661', 'd': '\U00002662', 'c': '\U00002663'}

    def __str__(self):
        return self._rank + self.suits[self._suit]

    def __eq__(self, other):
        if self._rank == other._rank:
            return self._suit == other._suit
        else:        
            return (self._rank == other._rank and self._suit == other._suit) 

class Hand:
    def __init__(self):
        self.cards = {'s': LinkedList(), 'h': LinkedList(), 'd': LinkedList(), 'c': LinkedList()}

    def __str__(self):
        result = ''
        for suit in self.cards.values():
            result += str(suit)
        return result

    def __getitem__(self, item):
        return self.cards[item]

    def __len__(self):
        result = 0
        for suit in list(self.cards):
            result += len(self.cards[suit])

        return result

    def add(self, card):
        self.cards[card._suit].add(card)

    def get_most_common_suit(self):
        return max(list(self.cards), key = lambda x: len(self.cards[x]))

    # Returns a card included in the hand according to
    # the criteria contained in *args and None if the card
    # isn't in the hand. The tests show how *args must be used.
    def play(self, *args):
        suit,rank = None,None

        for i in args:
            if i in self.cards:
                suit = str(i)
            else :
                rank = str(i)
        if suit and rank :
            temp =  self.cards[suit].remove(Card(rank,suit))
        elif suit:
            temp = self.cards[suit].pop()
        else:
            for i in self.cards:
                temp = self.cards[i].remove(Card(rank,str(i)))
                if temp : return temp
        return temp
                
        
class Deck(LinkedList):
    def __init__(self, custom=False):
        super().__init__()
        if not custom:
            # for all suits
            for i in range(4):
                # for all ranks
                for j in range(13):
                    s = list(Card.suits)[i]
                    r = ''
                    if j == 0:
                        r = 'A'
                    elif j > 0 and j < 10:
                        r = str(j+1)
                    elif j == 10:
                        r = 'J'
                    elif j== 11:
                        r = 'Q'
                    elif j == 12:
                        r = 'K'
                    self.add(Card(r,s))

    def draw(self):
        return self.pop()

    def shuffle(self, cut_precision = 0.05):
        # Cutting the two decks in two
        center = len(self) / 2
        k = round(random.gauss(center, cut_precision*len(self)))
        
        current = self._head
        
        for i in range(k-1):
            current = current.next
        other_deck = current.next
        current.next = None
        current = self._head
    
        if random.uniform(0,1) < 0.5:  
            other_deck, current = current, other_deck
            
        self._head = current
     
        while other_deck:
            temp = current.next
            current.next = other_deck
            other_deck = temp
            current = current.next
        
class Player():
    def __init__(self, name, strategy='naive'):
        self.name = name
        self.score = 8
        self.hand = Hand()
        self.strategy = strategy

    def __str__(self):
        return self.name

    # This function must modify the player's hand,
    # the discard pile, and the game's declared_suit 
    # attribute. No other variables must be changed.
    # The player's strategy can ONLY be based
    # on his own cards, the discard pile, and
    # the number of cards his opponents hold.
    def play(self, game):
        if(self.strategy == 'naive'):
            top_card = game.discard_pile.peek()
            if game.draw_count != 0 and game.declared_suit == '':
                temp = self.hand.play('2')
                if not temp:
                    temp = self.hand.play('Q','s')
            elif game.draw_count != 0 and game.declared_suit != '':
                temp = self.hand.play('2', game.declared_suit)
                if not temp and game.declared_suit == 's':
                    temp = self.hand.play('Q','s')
            elif game.declared_suit != '' :
                temp = self.hand.play(game.declared_suit)
                if temp and temp._rank == str(self.score):
                    temp2 = self.hand.play(game.declared_suit)
                    if temp2:
                        self.hand.add(temp)
                        temp = temp2
                elif not temp:
                    temp = self.hand.play(str(self.score))

            else :
                temp = self.hand.play(str(top_card._suit))
                if not temp:
                    temp = self.hand.play(top_card._rank)
                    if not temp:
                        temp = self.hand.play(str(self.score))
                        if self.score == 1 :
                            temp = self.hand.play('A')
                elif temp._rank == str(self.score):
                    temp2 =  self.hand.play(str(top_card._suit))
                    if temp2:
                        self.hand.add(temp)
                        temp = temp2
                    else:
                        temp2 = self.hand.play(str(top_card._rank))
                        if temp2:
                            self.hand.add(temp)
                            temp = temp2
                        else:
                            temp2 = self.hand.play(str(self.score))
                            if temp2:
                                if temp2._suit > temp._suit:
                                    self.hand.add(temp)
                                    temp = temp2
                                else:
                                    self.hand.add(temp2)

                        
            if temp :
                game.discard_pile.add(temp)
                if temp._rank == str(self.score) or (temp._rank == 'A' and self.score == 1):
                    game.declared_suit = self.hand.get_most_common_suit()
            return game
        else:
            # TO DO(?): Custom strategy (Bonus)
            pass

class Game:
    def __init__(self):
        self.players = CircularLinkedList()
  
        for i in range(1,5):
            self.players.append(Player('Player '+ str(i)))

        self.deck = Deck()
        self.discard_pile = LinkedList()

        self.draw_count = 0
        self.declared_suit = ''

    def __str__(self):
        result = '--------------------------------------------------\n'
        result += 'Deck: ' + str(self.deck) + '\n'
        result += 'Declared Suit: ' + str(self.declared_suit) + ', '
        result += 'Draw Count: ' + str(self.draw_count) + ', '
        result += 'Top Card: ' + str(self.discard_pile.peek()) + '\n'

        for player in self.players:
            result += str(player) + ': '
            result += 'Score: ' + str(player.score) + ', '
            result += str(player.hand) + '\n'
        return result
    

    # Puts all cards from discard pile except the 
    # top card back into the deck in reverse order
    # and shuffles it 7 times
    def reset_deck(self):
        temp = self.discard_pile.pop()
        for i in range(len(self.discard_pile)):
            self.deck.append(self.discard_pile.pop())
        for i in range(7):
            self.deck.shuffle()
        if temp:
            self.discard_pile.add(temp)

    # Safe way of drawing a card from the deck
    # that resets it if it is empty after card is drawn
    def draw_from_deck(self, num):
        h = LinkedList()
        while num !=0:
            if len(self.deck)> 1 :
                h.append(self.deck.draw())
                num -= 1
            else:
                h.append(self.deck.draw())
                num -= 1
                self.reset_deck()
        return h
        
        
            
    def start(self, debug=False):
        # Ordre dans lequel les joeurs gagnent la partie
        result = LinkedList()
        self.reset_deck()
        # Each player draws 8 cards
        for player in self.players:
            for i in range(8):
                player.hand.add(self.deck.draw())
        self.discard_pile.add(self.deck.draw())
        transcript = open('result.txt','w', encoding='utf-8')
        if debug:
            transcript = open('result_debug_test.txt','w', encoding='utf-8')
      
        while(not self.players.isEmpty()):
          
            if debug:
                transcript.write(str(self))

            # player plays turn
            player = self.players.peek()

            old_top_card = self.discard_pile.peek()
            
            self = player.play(self)
            new_top_card = game.discard_pile.peek()
            # Player didn't play a card => must draw from pile
            if new_top_card == old_top_card:
                if self.draw_count == 0 :
                    h = game.draw_from_deck(1)
                    player.hand.add(h.pop())
                    transcript.write(str(player) + ' draws 1 card' + '\n')
                else:
                    h = game.draw_from_deck(self.draw_count)
                    for i in range(len(h)):
                        player.hand.add(h.pop())
                    transcript.write(str(player) + ' draws '+ str(self.draw_count) + ' cards' + '\n')
                self.draw_count = 0
                old_top_card = None
            # Player played a card
            else:
                transcript.write(str(player) + ' plays ' + str(new_top_card)+'\n' )
                if new_top_card._rank == '2':
                    self.draw_count += 2
                elif new_top_card == Card('Q','s'):
                    self.draw_count += 5
                elif new_top_card._rank == 'A':
                    self.players.reverse()
                elif new_top_card._rank == 'J':
                    self.players.next()
                if new_top_card._rank != str(player.score):##attention si score == 1
                    if not (new_top_card._rank == 'A' and player.score == 1):
                        self.declared_suit = ''
                
                
            # Handling player change
            # Player has finished the game
            if len(player.hand) == 0 and player.score == 1:
                
                if len(result) < 2:
                    transcript.write(str(player) +' finishes in position ' + str(len(result)+1) +'\n')
                    result.append(str(self.players.pop()))
                else :
                    transcript.write((str(player) +   ' finishes in position ' + str(len(result)+1) +'\n'))
                    result.append(str(self.players.pop()))
                    last = str(self.players.pop())
                    result.append(last) 
                    transcript.write(last +   ' finishes last')
            else:
                # Player is out of cards to play
                if len(player.hand) == 0:
                    player.score -= 1
                    transcript.write(str(player) +' is out of cards to play! ' +
                                     str(player) + ' draws ' + str(player.score) + ' cards' +'\n' )
                    for i in range(player.score):
                        player.hand.add(game.draw_from_deck(1).pop())
                # Player has a single card left to play
                elif len(player.hand) == 1:
                     transcript.write('*Knock, knock* - '+str(player) +' has a single card left!' + '\n')
                self.players.next()
        return result
```
